```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('iris.csv' , encoding = 'utf-8')
```

```python
print(df.head(5))
```

```python
y = df.iloc[: , -1].values
x = df.iloc[: ,1:-1].values
```

```python
#applying label encoding to the y
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
#print(y)
```

```python
#spltiing the dataset into test,train 
from sklearn.model_selection import train_test_split
x_train ,x_test , y_train , y_test = train_test_split(x,y,test_size = .2 , random_state = 0)
#print(len(x_train))
#print(len(df))
```

```python
#impoting the decision tree classifier 
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy' , random_state = 0)
classifier.fit(x_train , y_train )
```

```python
#pridicting the random value
a = classifier.predict([[5.9,3,5.1,1.8]])
```

```python
print(le.inverse_transform(a))
```

```python
#now predicting the x_test value
y_pred = classifier.predict(x_test)
#printing the real value and predicted value side by side
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#finding the accuracy 
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```
